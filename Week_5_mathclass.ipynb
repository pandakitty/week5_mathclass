```python
# Using the Right Riemann Sum to Find the AUC


## (a) sketch a graph o ff and the rectangles that make up the Right Riemann sum R4 on the interval (0,2))
# import libraries
import numpy as np
from numpy import sqrt
import matplotlib.pyplot as plt

f= lambda x: sqrt(4-x**2)
a = 0; b = 2; N = 4
n = 4
x=np.linspace(a,b,N+1)
y=f(x)

X = np.linspace(a,b,n*N+1)
Y = f(X)

plt.figure(figsize=(15,6))
plt.subplot(1,3,1)
plt.plot(X,Y,'m')
x_right = x[1:]
y_right = y[1:]
plt.plot(x_right, y_right, 'm.', markersize=11)
plt.bar(x_right, y_right, width=-(b-a)/N,alpha=0.2, align= 'edge', edgecolor= 'm')
plt.title('Right-end Riemann Sum, N = {}'.format(N))
```

```python
#3(c)) Calculate appropriate right Riemann sums to find the AOC y = f(x) on the interval [0,2]. Does the value of this area have a special name?

import numpy as np 
def num_int(f,a,b,n): 
    dx=(b-a)/n 
    x=np.arange(a + dx,b,step=dx)
    y=f(x) 
    return y.sum()*dx 

def rational_func(x): 
    return sqrt(4-x**2)
print(num_int(rational_func,0,2,10000000))
print("The value's special name is Pi")
```
